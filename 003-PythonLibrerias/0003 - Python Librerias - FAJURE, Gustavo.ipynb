### Ejercicio 2

```
import pandas as pd

datos = {
    'Estudiante': ['Ana', 'Juan', 'Carla', 'Pedro', 'Lucia'],
    'Nota': [8, 6, 9, 5, 7]
}
df = pd.DataFrame (datos)

print (df)
```

```
promedio = df['Nota'].mean()
print (f"El promedio de las notas es: {promedio}")

nota_maxima = df['Nota'].max()
print (f"La nota máxima es: {nota_maxima}")

nota_minima = df['Nota'].min()
print (f"La nota mínima es: {nota_minima}")
```
